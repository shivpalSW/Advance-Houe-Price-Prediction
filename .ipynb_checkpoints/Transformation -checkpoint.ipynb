```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns',None)
```

```python
df = pd.read_csv('Ames_missing_values_removed')
```

```python
df.head()
```

```python
df[['Year Built', 'Year Remod/Add', 'Garage Yr Blt','Yr Sold']]
```

```python
df['MS SubClass'] = df['MS SubClass'].astype('O')
```

```python
Numerical_features = df.select_dtypes(exclude='O')
Numerical_features
```

```python
Categorical_features = df.select_dtypes(include='O')
Categorical_features
```

```python
Categorical_features = pd.get_dummies(Categorical_features,drop_first=True)
```

```python
Categorical_features
```

```python
Numerical_features
```

```python
for i in Numerical_features.columns:
    sns.histplot(data=Numerical_features,x=i)
    plt.show()
```

```python
Continous_features = []
for feature in Numerical_features.columns:
    if len(Numerical_features[feature].unique()) > 25 and feature not in ['Year Built', 'Year Remod/Add', 'Garage Yr Blt', 'Yr Sold']:
        Continous_features.append(feature)
```

```python
for feature in Continous_features:
    if 0 in Numerical_features[feature].unique():
        pass
    else:
        Numerical_features[feature] = np.log(Numerical_features[feature])
```

```python
for i in Continous_features:
    if 0 in df[i].unique():
        pass
    else:
        sns.histplot(data = df[Continous_features],x=i)
        plt.show()
```

```python
for i in Continous_features:
    if 0 in df[i].unique():
        pass
    else:
        print(i)
```
